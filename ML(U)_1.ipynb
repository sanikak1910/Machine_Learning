Encoding Categorical Values (Independent Variable)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),['person_gender','person_education','person_home_ownership','loan_intent','previous_loan_defaults_on_file'])], remainder='passthrough')
```

```python
x = np.array(ct.fit_transform(X))
x
```

```python
pd.DataFrame(x)
```

Encoding Categorical Variable(Dependent Variable) Output label

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le =LabelEncoder()
y = le.fit_transform(Y)
```

```python
y
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

**Feature Scaling**

Feature Scaling should be done after train and test split to prevent data leakage.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,19:] = sc.fit_transform(X_train[:,19:])
X_test[:,19:] = sc.fit_transform(X_test[:,19:])
```
