```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/Salary_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,1:-1]
```

```python
X
```

```python
Y = df.iloc[:,-1]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
y_train
```

Linear Regression

Training the simple linear regression on training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Predicting the test set result

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

Visualizing the training set

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Visualizing Testing data

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

# **Multiple Linear Regression**

```python
startup = pd.read_csv('/content/50_Startups_dataset.csv')
```

```python
startup
```

```python
X = startup.iloc[:,1:-1]
Y = startup.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
X.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
multi_reg = LinearRegression()
multi_reg.fit(X_train, y_train)
```

```python
y_pred = multi_reg.predict(X_test)
```

```python
y_pred
```

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.values.reshape(len(y_test),1)),1))
```

# **Polynomial Regression**

```python
salaries = pd.read_csv('/content/position_salaries.csv')
```

```python
salaries
```

```python
X = salaries.iloc[:,1:-1].values
Y = salaries.iloc[:,-1].values
```

```python
X
```

```python
Y
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regression = PolynomialFeatures(degree=4)
x_poly = poly_regression.fit_transform(X)
lin_reg = LinearRegression()
lin_reg.fit(x_poly,Y)
```

```python
plt.scatter(X, Y, color='red')
plt.plot(X, linear_reg.predict(X),color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X, Y, color='red')
plt.plot(X, lin_reg.predict(poly_regression.fit_transform(X)),color='blue')
plt.title('polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#For better resolution and smooth curve
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))

plt.scatter(X, Y, color='red')
plt.plot(X_grid, lin_reg.predict(poly_regression.fit_transform(X_grid)),color='blue')
plt.title('polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

Predicting on new data with Linear Regression

```python
linear_reg.predict([[6.5]])
```

Predicting on new data with Polynomial Regression

```python
lin_reg.predict(poly_regression.fit_transform([[6.5]]))
```
