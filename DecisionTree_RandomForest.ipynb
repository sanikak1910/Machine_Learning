```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
salaries = pd.read_csv('position_salaries.csv')
```

```python
X = salaries.iloc[:, 1:-1].values
y = salaries.iloc[:, -1].values
```

```python
X
```

```python
y
```

No feature scaling as they perform succesive splits


Training Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

```python
y_pred = regressor.predict([[6.5]])
```

```python
y_pred
```

Visualizing Decision Tree Results

```python
X_grid = np.arange(min(X), max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))

plt.scatter(X,y,color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title('Decision tree')
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```

# **Random Forest**

High Dimensional data

```python
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor(n_estimators=10, random_state=0)
random_forest.fit(X,y)
```

```python
y_pred_rf = random_forest.predict([[6.5]])
y_pred_rf
```

```python
X_grid = np.arange(min(X), max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))

plt.scatter(X,y,color='red')
plt.plot(X_grid,random_forest.predict(X_grid),color='blue')
plt.title('Decision tree')
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```

```python
from sklearn.metrics import r2_score
y_pred_rf_all = random_forest.predict(X)
r2_score(y, y_pred_rf_all)
```
